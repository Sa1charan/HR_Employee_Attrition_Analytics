**Plot a Correlation map for all numeric variables**

```python
f,ax = plt.subplots(figsize=(20, 20))
sns.heatmap(HR.corr(),annot=True, linewidths=5, fmt= '.1f' ,ax=ax)
```

```python
#count of people doing overtime
#sns.countplot(HR["OverTime"])
#fig = plt.gcf()
#fig.set_size_inches(10,10)
#plt.title("OverTime")
sns.countplot(data=HR, x="OverTime")
plt.title("OverTime")
plt.show()
```

The count of people doing overtime < the count of people not doing overtime

```python
#count of people based on their marital status
sns.countplot(data=HR, x="MaritalStatus")
plt.title("MaritalStatus")
plt.show()
```

There are many married people than single and divorsed.

```python
sns.countplot(data=HR, x="JobRole")
fig = plt.gcf()
fig.set_size_inches(20,14)
plt.title("JobRole")
plt.show()
```

These are the different job roles of the people.

```python
sns.countplot(data=HR, x="Gender")
fig = plt.gcf()
fig.set_size_inches(8,6)
plt.title("Gender")
plt.show()
```

```python
#analyzing the age group of the people doing overtime
sns.catplot(x='OverTime', y='Age', kind='swarm', data=HR)
```

```python
#count of people based on their total working years
#lt.subplots(figsize=(15,5))
#ns.countplot(HR.TotalWorkingYears)
sns.countplot(data=HR, x="TotalWorkingYears")
fig = plt.gcf()
fig.set_size_inches(15,5)
plt.title("")
plt.show()
```

The highest count of people are 200 with an experience of 10 years.

```python
#count of people based on their education
sns.countplot(data=HR, x="Education")
fig = plt.gcf()
fig.set_size_inches(15,5)
plt.title("")
plt.show()
```

```python
#count of people based on the distance from home they are working
sns.countplot(data=HR, x="DistanceFromHome")
fig = plt.gcf()
fig.set_size_inches(15,5)
plt.title("")
plt.show()
```
